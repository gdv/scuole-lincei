```python
import pandas as pd
```

```python
parole = pd.read_csv("https://github.com/napolux/paroleitaliane/raw/master/paroleitaliane/60000_parole_italiane.txt",
                     names = ['parola'])
parole.head()
```

```python
def canonica(s):
    return ''.join(sorted(s))
```

```python
canonica("banana")
```

```python
parole['firma'] = parole['parola'].apply(canonica)
parole
```

```python
conteggio = parole.groupby("firma").size()
```

```python
conteggio.head()
```

```python
ordinati = conteggio.sort_values(ascending = False).head(1)
ordinati
```

```python
ordinati.index
```

```python
list(ordinati.index)
```

```python
anagrammi = parole[parole['firma'].isin(list(ordinati.index))]
anagrammi
```

```python
anagrammi.sort_values('firma')
```
